```python
import pandas as pd
```

```python
#Normalizar os dados
dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dados/dados_normalizar.csv', sep =';')
#Segmentar dados escalares de dados categóricos
dados_num = dados.drop(columns=['sexo'])
dados_cat = dados['sexo']

#Normalzar os dados categóricos (coluna sexo)
dados_cat_normalizado = pd.get_dummies(data=dados_cat,prefix_sep='_', prefix='sexo')
dados_cat_normalizado

#Normalizar os dados numéricos
#1 Converter os dados numericos em ndarray
from sklearn import preprocessing
normalizador = preprocessing.MinMaxScaler()
modelo_normalizador = normalizador.fit(dados_num) #treinar o modelo que normalizará todos os dados relativos a IA aqui desenvolvida
dados_num_normalizado = modelo_normalizador.fit_transform(dados_num)

#Recompor os dados normalizados em um objeto do tipo DataFrame
dados_final = pd.DataFrame(data = dados_num_normalizado, columns = ['idade', 'altura', 'Peso'])

#Juntar os dados numéricos normalizados com os dados categoricos normalizados
dados_final = dados_final.join(dados_cat_normalizado, how = 'left')
dados_final
```

```python
#Salvar o normalizador para uso posterior
from pickle import dump
dump(modelo_normalizador,open('/content/drive/MyDrive/Colab Notebooks/dados/modelo_normalizador_1.pkl', 'wb'))
```

```python
#Normalizar os dados de um novo paciente (instância desconhecida)
#Carregar o modelo normalizador
from pickle import load
modelo_normalizador  = load(open('/content/drive/MyDrive/Colab Notebooks/dados/modelo_normalizador_1.pkl', 'rb'))

#Nova instância dever uma lista de dados
novo_paciente = [[51,1.86,84]]
novo_paciente_normalizado = modelo_normalizador.transform(novo_paciente)
novo_paciente_normalizado
```
